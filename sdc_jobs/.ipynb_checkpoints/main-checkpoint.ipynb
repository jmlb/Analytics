```python
class Indeed_JobScrap():
    
    def __init__(self, db_fname, overwrite_db):
        self.overwrite_db = overwrite_db
        self.db_fname = db_fname
        self.job_items = ["jobtitle", "date", "company", "country", "state", 
                          "longitude", "latitude"]
        self.tracker = [] #tracker how many consecutives job_response ar null
        self.thres = 3 #
        self.listing = {}
    
    
    def connect_db(self):
        sql_files = glob.glob('*.db')
        if (self.db_fname in sql_files) or (overwrite_db == False):
            self.conn = sqlite3.connect(self.db_fname)
            self.c = self.conn.cursor()
        else:
            #open connection to local file
            self.conn = sqlite3.connect(self.db_fname)
            self.c = self.conn.cursor()
            # Create table
            try:
                self.c.execute("CREATE TABLE jobs (jobkey, jobtitle, company, date, country, state, longitude, latitude))".format("jobs"))
                print('Created Table: {} in database {}'.format("jobs", self.dbname) )
                # Save (commit) the changes
                self.conn.commit()
            except:
                print('Unable to create table')


                    
    def save_data2db(self, job_entry):
        #check if job in db
        sql_select = "SELECT jobs.jobkey FROM jobs WHERE jobs.jobkey = '{}';".format(job_entry['jobkey'])
        flag_job_indb = False
        flag_error = False
        try:
            self.c.execute(sql_select)
            result = self.c.fetchall()
            if len(result) != 0:
                flag_job_indb = True
        except:
            # Insert a row of data
            print('Unable to execute SQL query: SELECT')
            flag_error = True
                
        if flag_job_indb == False:
            item = self.clean_award_data(item)
            sql_insert = "INSERT INTO jobs VALUES ('{}', '{}', '{}', '{}', '{}', '{}', '{}', '{}');".format(job_entry['jobkey'], 
            job_entry['jobkey'], job_entry['jobtitle'], job_entry['company'],
            job_entry['date'], job_entry['country'], job_entry['state'],
            job_entry['longitude'], job_entry['latitude'])
            try:
                self.c.execute(sql_insert)
                self.conn.commit()
                return True
            except:
                print('Unable to save data on database')
                return False


    def send_query(self):
        query = ""
        if len(self.keywords) > 0:
            #build query
            for item in keywords:
                query = query +"+"+item
        
            params = {
                      'q' : query,
                      # 'l' : "austin",
                      'start': start,
                      'limit': limit, #maximum nbr of results returned
                      'userip' : "1.2.3.4",
                      'useragent' : "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_8_2)"
                     }

            return client.search(**params)

    def sanitize_text(self, text):
        try:
            text = text.decode('unicode_escape').encode('ascii', 'ignore') 
        except:
            pass

        #Remove special characters
        special_chars = "!@#$%^&*()[]{};:,./<>?\|`~=_"
        for special_char in special_chars:
            clean_text = text.replace(special_char, " ")
        #remove newline
        clean_text = " ".join(clean_text.split())

        clean_text = re.sub("[^a-zA-Z.+3]"," ", clean_text)  # Now get rid of any terms that aren't words (include 3 for d3.js)
                                                # Also include + for C++
        return clean_text
    
    def snif_kwords(self, jobkey):
        url = 'http://www.indeed.com/viewjob?jk='+str(jobkey)
        flag_err = False
        try:
            site = urllib.request.urlopen(url).read() # Connect to the job posting
            soup_obj = BeautifulSoup(site, "lxml")
        except:
            print('page not found')
            flag_error = True
   
        if flag_error == False:
            spans = soup_obj.find_all('span', attrs={'id':"job_summary"})
            if len(spans) > 0:
                #convert soup object to text
                text = spans[0].get_text()
            #Cleaning text
            clean_text = self.sanitize_text(text)
            text_split = clean_text.split()
            return text_split
    
    
    
    def record(self, keywords, start, limit):
        while len(self.tracker) < self.thres:
            query_response = self.send_query(keywords, start, limit)
            n_entry = 0
            for entry in query_response['results']:
                n_entry += 1
                jobkey = entry['jobkey']
                job_detail = {}
                #Make a query on this jobkey
                entry_response = client.jobs(jobkeys = (jobkey, ))
                if len(entry_response['results']) > 0:
                    #reset tracker
                    self.tracker = []
                    entry_response = entry_response['results'][0]
                    ##JOB DETAILS  ########
                    for job_item in self.job_items:
                        if job_item in entry_response:
                            job_detail[job_item] = entry_response[item]
                        else:
                            job_detail[item] = ""
                    job_detail['jobkey'] = jobkey
                    
                    self.save_data2db(job_detail)       
                            
                    time.sleep(1)
                    print('Job# {} | Jobkey {}'.format(i, jobkey))
                    print(self.listing)
                    print()
                    i += 1
                    
                else:
                    self.tracker.append(1)
                    print("empty")
            
            #check consistency n_entry and limit
            start = start + limit
                
        
```
